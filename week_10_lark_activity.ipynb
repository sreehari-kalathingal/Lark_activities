# 8. Check the outliers and handle outliers in performance score using Percentiles.


```python
plt.figure(figsize=(7,3))
sns.boxplot(x=auto['performance_score'])
plt.show()
```

```python
Q1=np.percentile(auto['performance_score'], 25, interpolation='midpoint')
Q2=np.percentile(auto['performance_score'], 50, interpolation='midpoint')
Q3=np.percentile(auto['performance_score'], 75, interpolation='midpoint')
IQR = Q3-Q1
low_limit = Q1-1.5*IQR
up_limit = Q3+1.5*IQR
print('Q1 : ',Q1,'\n'
     'Q2 : ',Q2,'\n'
     'Q3 : ',Q3)
print('IQR : ',IQR)
print('lower-limit : ',low_limit)
print('upper-limit : ',up_limit)
```

```python
outlier=[]
for x in auto['performance_score']:
    if ((x<low_limit) or (x>up_limit)):
        outlier.append(x)
print('outlier values are: ',outlier)
```

```python
ind1 = auto['performance_score']<low_limit
auto.loc[ind1].index
```
