# 2. Check whether female and male students scored the same marks.


```python
import scipy.stats as stats
stats.ttest_ind(data['Total Marks'][data['Gender'] == 'male'],
                 data['Total Marks'][data['Gender'] == 'female'])
```

solution

take Null Hypothesis Ho as 'the marks scored by male and female students are not same'
   
   df=n-1 = 28-1 = 27
   level of significance is 5% = 0.05
   
  then the table value is 2.052
  
  the calculated value is -0.694
  
  Here, |t score|=0.694 < 2.052 then accept the null hypothesis 
  
  Also, p value=0.49>0.05 implies accept Ho
  
  which shows that the marks of female and male students were not same


# 3. Whether test preparation helps the students?


```python
import scipy.stats as stats
stats.ttest_ind(data['Total Marks'][data['Test preparation'] == 'none'],
                 data['Total Marks'][data['Test preparation'] == 'completed'])
```

solution

lets take the Null Hypothesis  Ho as 'the Test preparation does not helps the students'
then the alternative hypothesis H1 is 'the test preparation helps the students'

df=n-1 = 28-1 =27
level of significance =5% = 0.05
the corresponding table value = 2.052

calculated value = -3.614
Here ,|t score|=3.614 > 2.052 , then reject Ho and accept alternative hypothesis

Also,p value = 0.00126 < 0.05 implies reject Ho and accept H1 

which shows that the test preparation helps the students
